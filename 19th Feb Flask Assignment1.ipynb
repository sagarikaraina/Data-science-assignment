## Q3. What is App routing in Flask? Why do we use app routes?

* App routing in Flask is the process of mapping URLs to Python functions in a Flask application.
* App Routing makes it simple to handle web applications or API's because each endpoint is handled by the distinct funtion.
* App routing in Flask is a powerful and flexible tool that allows to create complex web applications with multiple endpoints and dynamic URL patterns.Without app routing, it would be difficult to create an organized and easily navigable web application that can handle a large number of requests and respond to them quickly and accurately.

## Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
Company Name: ABC Corporation

Location: India

Contact Detail: 999-999-9999

```python
from flask import Flask

app = Flask(__name__)                      #object of flask class

@app.route("/welcome")                     #decorator to access function through url
def welcome_pg():                          #Function to be accessed by outside world
    return "Welcome to ABC Corporation"

@app.route("/")                        #decorator to access function through url
def company_details():                 #Function to be accessed by outside world
    return "<p>Company Name: ABC Corporation</p> <p>Location: India</p> <p>Contact Detail: 999-999-9999</p>"

if __name__=="__main__":                #in main function run obj app with run func and host as parameter to run the srver
    app.run(host="0.0.0.0",port=8000)   #turns this lab to develelopment server
```

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

* The url_for() function in Flask is used to build URLs for a given function name. This function generates a URL based on the specified endpoint and any arguments that are provided.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'

@app.route('/user/<username>')
def show_user_profile(username):
    return 'User %s' % username

@app.route('/post/<int:post_id>')
def show_post(post_id):
    return 'Post %d' % post_id

with app.test_request_context():
    print(url_for('index')) # Output: /
    print(url_for('show_user_profile', username='John Doe')) # Output: /user/John%20Doe
    print(url_for('show_post', post_id=1)) # Output: /post/1
```
